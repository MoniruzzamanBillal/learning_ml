## 🔢 Section 1: NumPy Basics

**Why NumPy instead of regular Python lists?**

Regular lists are slow for math because Python loops over each item one by one.
NumPy arrays do math on the *whole array at once* — much faster, and the syntax is cleaner.

Think of it like this: a JS array needs `.map()` to do math on every item. NumPy just lets you write the math directly.

```python
import numpy as np

# Likes for 8 posts (regular Python list)
likes_list = [4823, 921, 15200, 632, 2100, 3400, 8900, 5600]

# Convert to a NumPy array
likes = np.array(likes_list)

print(f"Type: {type(likes)}")
print(f"Array: {likes}")
print(f"Shape: {likes.shape}")   # (8,) means 8 elements, 1 dimension
```

```python
# ✅ The power of NumPy — math on the whole array at once

# With a normal Python list, you'd need:
doubled_list = [x * 2 for x in likes_list]   # list comprehension required

# With NumPy, just write the math directly:
doubled_array = likes * 2
print(f"Doubled: {doubled_array}")

# Add 100 bonus likes to every post
bonus = likes + 100
print(f"With bonus: {bonus}")

# Built-in statistics — no manual loops needed!
print(f"\nMean (average): {np.mean(likes):.1f}")
print(f"Max:             {np.max(likes)}")
print(f"Min:             {np.min(likes)}")
print(f"Sum:             {np.sum(likes)}")
print(f"Standard Dev:    {np.std(likes):.1f}")   # how spread out the data is
print(f"Median:          {np.median(likes):.1f}")  # middle value
```

```python
# ✅ Boolean filtering — extremely common in ML data cleaning

comments = np.array([312, 445, 870, 98, 180, 560, 430, 920])
shares = np.array([89, 230, 1200, 45, 60, 890, 670, 340])

# Find which posts have likes > 3000 (returns True/False array)
mask = likes > 3000
print(f"Mask (True/False): {mask}")

# Use the mask to filter the actual data
high_performing_likes = likes[mask]
print(f"High performing posts (likes): {high_performing_likes}")

# This is called 'boolean indexing' — you'll use this CONSTANTLY in ML
# Combine multiple arrays into a calculation
total_engagement = likes + comments + shares
print(f"\nTotal engagement per post: {total_engagement}")
print(f"Average engagement: {np.mean(total_engagement):.1f}")
```

## 🐼 Section 2: Pandas — Working With Real Datasets

Pandas introduces two key structures:
- **Series** — like a single column (a 1D NumPy array with labels)
- **DataFrame** — like a full spreadsheet/table (multiple columns)

If you've used SQL or worked with Prisma models, think of a DataFrame like a database table you can query and transform in code.

```python
import pandas as pd

# Create a DataFrame from a list of dictionaries (just like Week 1!)
data = [
    {"platform": "Instagram", "likes": 4823,  "comments": 312, "shares": 89,   "category": "tech"},
    {"platform": "Twitter",   "likes": 921,   "comments": 445, "shares": 230,  "category": "news"},
    {"platform": "TikTok",    "likes": 15200, "comments": 870, "shares": 1200, "category": "tech"},
    {"platform": "LinkedIn",  "likes": 632,   "comments": 98,  "shares": 45,   "category": "career"},
    {"platform": "Instagram", "likes": 2100,  "comments": 180, "shares": 60,   "category": "tech"},
    {"platform": "Twitter",   "likes": 3400,  "comments": 560, "shares": 890,  "category": "news"},
    {"platform": "TikTok",    "likes": 8900,  "comments": 430, "shares": 670,  "category": "entertainment"},
    {"platform": "YouTube",   "likes": 5600,  "comments": 920, "shares": 340,  "category": "tech"},
    {"platform": "LinkedIn",  "likes": 1200,  "comments": 210, "shares": 88,   "category": "career"},
    {"platform": "YouTube",   "likes": 9800,  "comments": 1200,"shares": 560,  "category": "entertainment"},
]

df = pd.DataFrame(data)

# .head() shows first 5 rows — first thing you ALWAYS run on a new dataset
df.head()
```

```python
# ✅ Exploring a DataFrame — your first moves on any dataset

print("=== Shape (rows, columns) ===")
print(df.shape)

print("\n=== Column names ===")
print(df.columns.tolist())

print("\n=== Data types ===")
print(df.dtypes)

print("\n=== Quick statistics summary ===")
df.describe()
```

```python
# ✅ Selecting columns and rows

# Select a single column (returns a Series)
likes_column = df["likes"]
print("Likes column:")
print(likes_column)

print("\n--- Select multiple columns ---")
subset = df[["platform", "likes"]]
print(subset)

print("\n--- Select rows by condition (like SQL WHERE / Prisma findMany where) ---")
tech_posts = df[df["category"] == "tech"]
print(tech_posts)
```

```python
# ✅ Creating new columns (feature engineering — a core ML skill!)

df["engagement"] = df["likes"] + df["comments"] + df["shares"]

# Conditional column — like a SQL CASE WHEN
df["is_viral"] = df["engagement"] >= 10000

df.head(10)
```

```python
# ✅ Sorting — like ORDER BY in SQL

top_posts = df.sort_values("engagement", ascending=False)
print("=== Posts ranked by engagement ===")
print(top_posts[["platform", "category", "engagement", "is_viral"]])
```

```python
# ✅ Group By — exactly like SQL GROUP BY / Prisma groupBy
# This is one of the MOST used Pandas operations

# Average engagement per platform
platform_avg = df.groupby("platform")["engagement"].mean().sort_values(ascending=False)
print("=== Average Engagement by Platform ===")
print(platform_avg)

print("\n=== Multiple aggregations at once ===")
platform_summary = df.groupby("platform").agg(
    avg_likes=("likes", "mean"),
    total_engagement=("engagement", "sum"),
    post_count=("platform", "count")
).sort_values("total_engagement", ascending=False)

print(platform_summary)
```

## 🕳️ Section 3: Handling Missing Data

Real-world datasets are messy. Some values will be missing (`NaN`). Knowing how to handle this is a critical, constantly-used ML skill.

```python
import numpy as np

# A messier, more realistic dataset with some missing values
messy_data = [
    {"platform": "Instagram", "likes": 4823,  "comments": 312,  "shares": 89},
    {"platform": "Twitter",   "likes": np.nan,"comments": 445,  "shares": 230},   # missing likes
    {"platform": "TikTok",    "likes": 15200, "comments": np.nan,"shares": 1200}, # missing comments
    {"platform": "LinkedIn",  "likes": 632,   "comments": 98,   "shares": np.nan},# missing shares
    {"platform": "Instagram", "likes": 2100,  "comments": 180,  "shares": 60},
]

messy_df = pd.DataFrame(messy_data)
print("=== Original data with missing values ===")
print(messy_df)

print("\n=== Check for missing values ===")
print(messy_df.isnull().sum())
```

```python
# ✅ Two common strategies for missing data

# Strategy 1: Drop rows with any missing values
dropped_df = messy_df.dropna()
print("=== After dropping missing rows ===")
print(dropped_df)

# Strategy 2: Fill missing values (often better — keeps more data)
filled_df = messy_df.copy()
filled_df["likes"] = filled_df["likes"].fillna(filled_df["likes"].mean())
filled_df["comments"] = filled_df["comments"].fillna(filled_df["comments"].mean())
filled_df["shares"] = filled_df["shares"].fillna(filled_df["shares"].mean())

print("\n=== After filling missing values with column mean ===")
print(filled_df)
```

## 🏋️ Section 4: Mini Project — Full Dataset Analysis

Time to combine everything. We'll simulate loading a CSV (the way you'll do with real datasets from Kaggle) and run a full analysis.

```python
# ✅ Step 1: Create and save a CSV (simulates downloading a real dataset)

import pandas as pd
import numpy as np

np.random.seed(42)   # makes random numbers reproducible

platforms = ["Instagram", "Twitter", "TikTok", "LinkedIn", "YouTube"]
categories = ["tech", "news", "entertainment", "career", "lifestyle"]

n_posts = 200

dataset = pd.DataFrame({
    "post_id": range(1, n_posts + 1),
    "platform": np.random.choice(platforms, n_posts),
    "category": np.random.choice(categories, n_posts),
    "likes": np.random.randint(100, 20000, n_posts),
    "comments": np.random.randint(10, 2000, n_posts),
    "shares": np.random.randint(5, 1500, n_posts),
})

# Save to CSV — this is how real datasets get stored
dataset.to_csv("social_media_data.csv", index=False)
print("✅ Dataset saved as social_media_data.csv")
dataset.head()
```

```python
# ✅ Step 2: Load the CSV (this is what you'll do with EVERY real dataset)

df = pd.read_csv("social_media_data.csv")

print(f"Dataset shape: {df.shape[0]} rows, {df.shape[1]} columns")
df.head()
```

```python
# ✅ Step 3: Full Analysis Pipeline

# Feature engineering
df["engagement"] = df["likes"] + df["comments"] + df["shares"]
df["is_viral"] = df["engagement"] >= df["engagement"].quantile(0.90)  # top 10% = viral

print("="*60)
print("           SOCIAL MEDIA DATASET ANALYSIS REPORT")
print("="*60)

print(f"\n📊 Total posts analyzed: {len(df)}")
print(f"🔥 Viral posts: {df['is_viral'].sum()} ({df['is_viral'].mean()*100:.1f}%)")

print("\n--- Engagement by Platform ---")
platform_report = df.groupby("platform").agg(
    avg_engagement=("engagement", "mean"),
    total_posts=("post_id", "count"),
    viral_count=("is_viral", "sum")
).round(0).sort_values("avg_engagement", ascending=False)
print(platform_report)

print("\n--- Engagement by Category ---")
category_report = df.groupby("category")["engagement"].mean().round(0).sort_values(ascending=False)
print(category_report)

print("\n--- Top 5 Most Viral Posts ---")
top5 = df.sort_values("engagement", ascending=False).head(5)
print(top5[["post_id", "platform", "category", "engagement"]].to_string(index=False))

print("\n" + "="*60)
best_platform = platform_report.index[0]
best_category = category_report.index[0]
print(f"🏆 Best performing platform: {best_platform}")
print(f"🏆 Best performing category: {best_category}")
print("="*60)
```

```python
# ✅ Step 4: Quick Visualization (preview of Week 5-6!)
import matplotlib.pyplot as plt

platform_report["avg_engagement"].plot(kind="bar", figsize=(8,5), color="#6366f1")
plt.title("Average Engagement by Platform")
plt.ylabel("Average Engagement")
plt.xlabel("Platform")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
