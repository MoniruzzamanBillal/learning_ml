## 🔁 Section 0: GroupBy — Slow, Step-by-Step Breakdown

Let's rebuild your groupby intuition from scratch. The confusion usually comes from not seeing **what happens in between** `.groupby()` and the final result. We'll break every step apart.

**The mental model:** groupby = **Split → Apply → Combine**
1. **Split** the data into groups (e.g., one group per platform)
2. **Apply** a function to each group (e.g., calculate the mean)
3. **Combine** the results back into one table

This is *exactly* like SQL's `GROUP BY` + aggregate function, or Prisma's `groupBy()`.

````python
import pandas as pd

data = [
    {"platform": "Instagram", "likes": 4823,  "comments": 312},
    {"platform": "Twitter",   "likes": 921,   "comments": 445},
    {"platform": "Instagram", "likes": 2100,  "comments": 180},
    {"platform": "TikTok",    "likes": 15200, "comments": 870},
    {"platform": "Twitter",   "likes": 3400,  "comments": 560},
    {"platform": "TikTok",    "likes": 8900,  "comments": 430},
]
df = pd.DataFrame(data)
df
````

````python
# ✅ STEP 1: SPLIT — see the groups Pandas creates

grouped = df.groupby("platform")

# This alone doesn't show data — it's just the 'plan' for grouping
print(grouped)
print(type(grouped))

# Use .groups to literally see which rows belong to which group
print("\nGroup membership (row indices):")
print(grouped.groups)

# You can loop through each group manually to SEE the split happening
print("\n--- Manually inspecting each group ---")
for platform_name, group_data in grouped:
    print(f"\nGroup: {platform_name}")
    print(group_data)
````

````python
# ✅ STEP 2: APPLY — run a calculation on each group

# This is the part that confuses people. Let's go SLOW.

# Pick ONE column, then aggregate it
avg_likes_per_platform = df.groupby("platform")["likes"].mean()
print("Average likes per platform:")
print(avg_likes_per_platform)
print(f"\nType: {type(avg_likes_per_platform)}")  # It's a Series!
````

````python
# ✅ STEP 3: COMBINE — the result is automatically combined into one table

# Try different aggregate functions — same pattern every time:
# df.groupby("COLUMN_TO_GROUP_BY")["COLUMN_TO_CALCULATE"].FUNCTION()

print("SUM of likes per platform:")
print(df.groupby("platform")["likes"].sum())

print("\nCOUNT of posts per platform:")
print(df.groupby("platform")["likes"].count())

print("\nMAX likes per platform:")
print(df.groupby("platform")["likes"].max())

print("\nMULTIPLE columns at once:")
print(df.groupby("platform")[["likes", "comments"]].mean())
````

````python
# ✅ The .agg() method — multiple different calculations at once
# This is the version that looks scary but is just shorthand for everything above

summary = df.groupby("platform").agg(
    avg_likes=("likes", "mean"),      # name=( column, function )
    total_likes=("likes", "sum"),
    post_count=("likes", "count"),
    avg_comments=("comments", "mean")
)
print(summary)

# Read it like this:
# "Group rows by platform. For each group, calculate:
#   - avg_likes = mean of the likes column
#   - total_likes = sum of the likes column
#   - post_count = count of rows
#   - avg_comments = mean of the comments column"
````

## 📈 Section 1: Matplotlib Basics

`matplotlib.pyplot` is the standard charting library in Python. The pattern is almost always:

```python
plt.plot(...)        # or plt.bar(), plt.scatter(), etc.
plt.title("...")
plt.xlabel("...")
plt.ylabel("...")
plt.show()
```

````python
import matplotlib.pyplot as plt

# --- Line chart: engagement over time ---
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
engagement = [1200, 1900, 1500, 2400, 3200, 4100, 3800]

plt.figure(figsize=(8, 5))
plt.plot(days, engagement, marker="o", color="#6366f1", linewidth=2)
plt.title("Weekly Engagement Trend")
plt.xlabel("Day of Week")
plt.ylabel("Total Engagement")
plt.grid(True, alpha=0.3)
plt.show()
````

````python
# --- Bar chart: comparing platforms ---
platforms = ["Instagram", "Twitter", "TikTok", "LinkedIn", "YouTube"]
avg_engagement = [3200, 2100, 8900, 950, 4500]

plt.figure(figsize=(8, 5))
bars = plt.bar(platforms, avg_engagement, color="#10b981")
plt.title("Average Engagement by Platform")
plt.xlabel("Platform")
plt.ylabel("Avg Engagement")

# Add value labels on top of bars — looks more professional
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 50, f"{height:,.0f}",
             ha="center", fontsize=9)

plt.show()
````

````python
# --- Scatter plot: relationship between two variables ---
# Does more comments mean more shares? Let's check visually.

import numpy as np
np.random.seed(1)

comments = np.random.randint(50, 1000, 50)
shares = comments * 0.8 + np.random.randint(-100, 100, 50)  # roughly correlated

plt.figure(figsize=(8, 5))
plt.scatter(comments, shares, color="#f59e0b", alpha=0.6)
plt.title("Comments vs Shares")
plt.xlabel("Comments")
plt.ylabel("Shares")
plt.grid(True, alpha=0.3)
plt.show()

# 💡 Scatter plots are how you visually spot correlations BEFORE
# building a model — this is literally a preview of Linear Regression in Phase 2!
````

````python
# --- Pie chart: category distribution ---
categories = ["tech", "news", "entertainment", "career", "lifestyle"]
post_counts = [45, 30, 60, 20, 35]

plt.figure(figsize=(7, 7))
plt.pie(post_counts, labels=categories, autopct="%1.1f%%",
        colors=["#6366f1", "#10b981", "#f59e0b", "#ef4444", "#8b5cf6"])
plt.title("Post Distribution by Category")
plt.show()
````

````python
# --- Multiple charts side by side using subplots ---

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Left chart: bar
axes[0].bar(platforms, avg_engagement, color="#6366f1")
axes[0].set_title("Avg Engagement by Platform")
axes[0].set_ylabel("Engagement")
axes[0].tick_params(axis="x", rotation=30)

# Right chart: line
axes[1].plot(days, engagement, marker="o", color="#10b981")
axes[1].set_title("Weekly Engagement Trend")
axes[1].set_ylabel("Engagement")

plt.tight_layout()
plt.show()
````

## 🏋️ Section 2: Mini Project — Social Media Trends Dashboard

We'll generate a realistic dataset, run groupby analysis, and build a multi-chart dashboard — combining everything from Week 1 through Week 6.

````python
# ✅ Step 1: Generate the dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

platforms = ["Instagram", "Twitter", "TikTok", "LinkedIn", "YouTube"]
categories = ["tech", "news", "entertainment", "career", "lifestyle"]

n_posts = 300
df = pd.DataFrame({
    "post_id": range(1, n_posts + 1),
    "platform": np.random.choice(platforms, n_posts),
    "category": np.random.choice(categories, n_posts),
    "likes": np.random.randint(100, 20000, n_posts),
    "comments": np.random.randint(10, 2000, n_posts),
    "shares": np.random.randint(5, 1500, n_posts),
})
df["engagement"] = df["likes"] + df["comments"] + df["shares"]

df.head()
````

````python
# ✅ Step 2: GroupBy analysis (your practice pays off here!)

platform_summary = df.groupby("platform").agg(
    avg_engagement=("engagement", "mean"),
    post_count=("post_id", "count")
).sort_values("avg_engagement", ascending=False)

category_summary = df.groupby("category")["engagement"].mean().sort_values(ascending=False)

print(platform_summary)
print()
print(category_summary)
````

````python
# ✅ Step 3: Build a 4-panel dashboard

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle("📱 Social Media Analytics Dashboard", fontsize=16, fontweight="bold")

# Panel 1: Avg engagement by platform (bar)
axes[0, 0].bar(platform_summary.index, platform_summary["avg_engagement"], color="#6366f1")
axes[0, 0].set_title("Avg Engagement by Platform")
axes[0, 0].tick_params(axis="x", rotation=30)

# Panel 2: Post count by platform (bar)
axes[0, 1].bar(platform_summary.index, platform_summary["post_count"], color="#10b981")
axes[0, 1].set_title("Post Count by Platform")
axes[0, 1].tick_params(axis="x", rotation=30)

# Panel 3: Engagement by category (horizontal bar)
axes[1, 0].barh(category_summary.index, category_summary.values, color="#f59e0b")
axes[1, 0].set_title("Avg Engagement by Category")

# Panel 4: Likes vs Engagement scatter
axes[1, 1].scatter(df["likes"], df["engagement"], alpha=0.4, color="#ef4444")
axes[1, 1].set_title("Likes vs Total Engagement")
axes[1, 1].set_xlabel("Likes")
axes[1, 1].set_ylabel("Engagement")

plt.tight_layout()
plt.show()
````
