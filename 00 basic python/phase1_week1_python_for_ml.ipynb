## 📦 Section 1: Python vs JavaScript — Key Differences

You already know JS. Here's a quick mental map:

| JavaScript | Python |
|---|---|
| `const x = 5` | `x = 5` |
| `let name = "Ali"` | `name = "Ali"` |

## 📋 Section 2: Lists — Your Main Data Structure in ML

In JavaScript you use arrays `[]`. In Python, they're called **lists** — same thing, different name.

In ML, almost all your data lives in lists (or list-like structures). Get very comfortable with them.

````python
# ✅ SECTION 2: Lists

# Social media platforms and their monthly active users (in millions)
platforms = ["Facebook", "YouTube", "Instagram", "TikTok", "Twitter", "LinkedIn", "Snapchat"]
users_millions = [2900, 2500, 2000, 1500, 550, 900, 750]

# Accessing elements — same as JS
print(f"First platform: {platforms[0]}")
print(f"Last platform: {platforms[-1]}")   # -1 = last element (Python trick!)

# Length
print(f"Total platforms: {len(platforms)}")

# Slicing — get a portion of a list  (JS: .slice())
top_3 = platforms[0:3]
print(f"Top 3 platforms: {top_3}")

# Adding to a list  (JS: .push())
platforms.append("Pinterest")
users_millions.append(450)
print(f"Updated platforms: {platforms}")
````

````python
# ✅ Loops in Python
# JS: for (let i = 0; i < arr.length; i++)
# Python: for item in list  ← much cleaner!

print("=== All Platforms ===")
for platform in platforms:
    print(f"  - {platform}")

print()

# Loop with index — use enumerate() (like JS .forEach with index)
print("=== Platform Rankings ===")
for index, platform in enumerate(platforms):
    print(f"  #{index + 1}: {platform}")
````

## 🔥 Section 3: List Comprehensions — THE Most Important Python Skill for ML

This is used **everywhere** in ML code. Master this.

```python
# JS equivalent:
# arr.filter(x => x > 1000).map(x => x * 2)

# Python list comprehension:
# [expression for item in list if condition]
```

````python
# ✅ SECTION 3: List Comprehensions

users_millions = [2900, 2500, 2000, 1500, 550, 900, 750, 450]
platforms = ["Facebook", "YouTube", "Instagram", "TikTok", "Twitter", "LinkedIn", "Snapchat", "Pinterest"]

# --- Example 1: Filter ---
# Get platforms with more than 1000 million users

# JS way: users_millions.filter(u => u > 1000)
# Python way:
large_platforms_users = [u for u in users_millions if u > 1000]
print(f"Platforms with 1B+ users (numbers): {large_platforms_users}")

# --- Example 2: Filter with paired lists ---
large_platforms = [p for p, u in zip(platforms, users_millions) if u > 1000]
print(f"Platforms with 1B+ users (names): {large_platforms}")

# --- Example 3: Transform (map) ---
# Convert millions to billions
users_billions = [u / 1000 for u in users_millions]
print(f"Users in billions: {users_billions}")

# --- Example 4: Filter + Transform together ---
# Get names of small platforms (under 1B), in UPPERCASE
small_platforms = [p.upper() for p, u in zip(platforms, users_millions) if u < 1000]
print(f"Small platforms: {small_platforms}")
````


## 📊 Section 5: Basic Statistics — The Heart of ML

Every ML algorithm at its core is doing math on numbers. These basic stats are what you'll use constantly:

- **Mean (average)** — what's the center?
- **Max / Min** — what are the extremes?
- **Count** — how many samples?

We'll calculate these **manually first** (so you understand it), then using Python's built-in tools.

````python
# ✅ SECTION 5: Basic Statistics on Social Media Data

posts = [
    {"id": 1, "platform": "Instagram", "likes": 4823, "comments": 312, "shares": 89,  "category": "tech"},
    {"id": 2, "platform": "Twitter",   "likes": 921,  "comments": 445, "shares": 230, "category": "news"},
    {"id": 3, "platform": "TikTok",    "likes": 15200,"comments": 870, "shares": 1200,"category": "tech"},
    {"id": 4, "platform": "LinkedIn",  "likes": 632,  "comments": 98,  "shares": 45,  "category": "career"},
    {"id": 5, "platform": "Instagram", "likes": 2100, "comments": 180, "shares": 60,  "category": "tech"},
    {"id": 6, "platform": "Twitter",   "likes": 3400, "comments": 560, "shares": 890, "category": "news"},
    {"id": 7, "platform": "TikTok",    "likes": 8900, "comments": 430, "shares": 670, "category": "entertainment"},
    {"id": 8, "platform": "YouTube",   "likes": 5600, "comments": 920, "shares": 340, "category": "tech"},
]

# Extract likes
all_likes = [post["likes"] for post in posts]

# --- Manual calculations (understand the math!) ---
total_likes = sum(all_likes)
count = len(all_likes)
average_likes = total_likes / count
max_likes = max(all_likes)
min_likes = min(all_likes)

print("=== LIKES STATISTICS ===")
print(f"Total likes:    {total_likes:,}")
print(f"Post count:     {count}")
print(f"Average likes:  {average_likes:,.1f}")
print(f"Max likes:      {max_likes:,}")
print(f"Min likes:      {min_likes:,}")
````

````python
# ✅ Add Total Engagement to every post and find the most viral one

# Add engagement score to each post
for post in posts:
    post["engagement"] = post["likes"] + post["comments"] + post["shares"]

# Find most viral post
# In JS: posts.sort((a,b) => b.engagement - a.engagement)[0]
# Python:
most_viral = max(posts, key=lambda post: post["engagement"])

print("=== MOST VIRAL POST ===")
print(f"Platform:   {most_viral['platform']}")
print(f"Category:   {most_viral['category']}")
print(f"Likes:      {most_viral['likes']:,}")
print(f"Engagement: {most_viral['engagement']:,}")

# Sort all posts by engagement (high to low)
# In JS: posts.sort((a,b) => b.engagement - a.engagement)
sorted_posts = sorted(posts, key=lambda p: p["engagement"], reverse=True)

print("\n=== POSTS RANKED BY ENGAGEMENT ===")
for rank, post in enumerate(sorted_posts):
    print(f"  #{rank+1} {post['platform']:12} | Engagement: {post['engagement']:,}")
````

## ⚙️ Section 6: Functions in Python

Same concept as JS, slightly different syntax.

```javascript
// JS
function calculateAverage(numbers) {
    return numbers.reduce((a, b) => a + b, 0) / numbers.length;
}
```
```python
# Python
def calculate_average(numbers):
    return sum(numbers) / len(numbers)
```

````python
# ✅ SECTION 6: Functions

def calculate_stats(values, label="Values"):
    """Calculate and print basic statistics for a list of numbers."""
    average = sum(values) / len(values)
    print(f"\n=== {label} Stats ===")
    print(f"  Count:   {len(values)}")
    print(f"  Average: {average:,.1f}")
    print(f"  Max:     {max(values):,}")
    print(f"  Min:     {min(values):,}")
    return average


def get_posts_by_platform(posts, platform_name):
    """Filter posts by a specific platform."""
    return [p for p in posts if p["platform"] == platform_name]


def engagement_rate(post):
    """Calculate engagement score for a single post."""
    return post["likes"] + (post["comments"] * 2) + (post["shares"] * 3)
    # Comments and shares weighted higher than likes — common in real ML features!


# --- Use the functions ---
posts = [
    {"id": 1, "platform": "Instagram", "likes": 4823, "comments": 312, "shares": 89,  "category": "tech"},
    {"id": 2, "platform": "Twitter",   "likes": 921,  "comments": 445, "shares": 230, "category": "news"},
    {"id": 3, "platform": "TikTok",    "likes": 15200,"comments": 870, "shares": 1200,"category": "tech"},
    {"id": 4, "platform": "LinkedIn",  "likes": 632,  "comments": 98,  "shares": 45,  "category": "career"},
    {"id": 5, "platform": "Instagram", "likes": 2100, "comments": 180, "shares": 60,  "category": "tech"},
    {"id": 6, "platform": "Twitter",   "likes": 3400, "comments": 560, "shares": 890, "category": "news"},
    {"id": 7, "platform": "TikTok",    "likes": 8900, "comments": 430, "shares": 670, "category": "entertainment"},
    {"id": 8, "platform": "YouTube",   "likes": 5600, "comments": 920, "shares": 340, "category": "tech"},
]

all_likes = [p["likes"] for p in posts]
avg = calculate_stats(all_likes, "Likes")

instagram_posts = get_posts_by_platform(posts, "Instagram")
print(f"\nInstagram posts: {len(instagram_posts)}")

# Add weighted engagement to each post
for post in posts:
    post["weighted_engagement"] = engagement_rate(post)

print("\n=== Weighted Engagement ===")
for post in sorted(posts, key=lambda p: p["weighted_engagement"], reverse=True):
    print(f"  {post['platform']:12} | Score: {post['weighted_engagement']:,}")
````

## 🏋️ Section 7: Mini Project — Social Media Analyzer

Now put everything together. This is your **Week 1-2 project.**

You'll build a small analyzer that:
1. Takes a dataset of posts
2. Calculates stats per platform
3. Identifies viral posts
4. Prints a clean report

> This is **exactly** the kind of thinking you'll use in real ML pipelines — load data, analyze it, extract insights.

````python
# ✅ MINI PROJECT: Social Media Engagement Analyzer

# --- Dataset ---
posts = [
    {"id": 1,  "platform": "Instagram", "likes": 4823,  "comments": 312, "shares": 89,   "category": "tech"},
    {"id": 2,  "platform": "Twitter",   "likes": 921,   "comments": 445, "shares": 230,  "category": "news"},
    {"id": 3,  "platform": "TikTok",    "likes": 15200, "comments": 870, "shares": 1200, "category": "tech"},
    {"id": 4,  "platform": "LinkedIn",  "likes": 632,   "comments": 98,  "shares": 45,   "category": "career"},
    {"id": 5,  "platform": "Instagram", "likes": 2100,  "comments": 180, "shares": 60,   "category": "tech"},
    {"id": 6,  "platform": "Twitter",   "likes": 3400,  "comments": 560, "shares": 890,  "category": "news"},
    {"id": 7,  "platform": "TikTok",    "likes": 8900,  "comments": 430, "shares": 670,  "category": "entertainment"},
    {"id": 8,  "platform": "YouTube",   "likes": 5600,  "comments": 920, "shares": 340,  "category": "tech"},
    {"id": 9,  "platform": "LinkedIn",  "likes": 1200,  "comments": 210, "shares": 88,   "category": "career"},
    {"id": 10, "platform": "YouTube",   "likes": 9800,  "comments": 1200,"shares": 560,  "category": "entertainment"},
]


# --- Step 1: Add engagement score to every post ---
VIRAL_THRESHOLD = 10000   # posts above this engagement are "viral"

for post in posts:
    post["engagement"] = post["likes"] + post["comments"] + post["shares"]
    post["is_viral"] = post["engagement"] >= VIRAL_THRESHOLD


# --- Step 2: Group stats by platform ---
def platform_stats(posts, platform):
    filtered = [p for p in posts if p["platform"] == platform]
    if not filtered:
        return None
    engagements = [p["engagement"] for p in filtered]
    return {
        "platform": platform,
        "post_count": len(filtered),
        "avg_engagement": sum(engagements) / len(engagements),
        "max_engagement": max(engagements),
        "viral_posts": len([p for p in filtered if p["is_viral"]])
    }


# --- Step 3: Get unique platforms ---
unique_platforms = list(set([p["platform"] for p in posts]))


# --- Step 4: Generate report ---
print("=" * 55)
print("       SOCIAL MEDIA ENGAGEMENT REPORT")
print("=" * 55)

all_stats = []
for platform in unique_platforms:
    stats = platform_stats(posts, platform)
    all_stats.append(stats)

# Sort by average engagement
all_stats.sort(key=lambda s: s["avg_engagement"], reverse=True)

for stats in all_stats:
    print(f"\n📱 {stats['platform']}")
    print(f"   Posts analyzed:   {stats['post_count']}")
    print(f"   Avg engagement:   {stats['avg_engagement']:,.0f}")
    print(f"   Best post:        {stats['max_engagement']:,}")
    print(f"   Viral posts:      {stats['viral_posts']}")

print("\n" + "=" * 55)
viral_posts = [p for p in posts if p["is_viral"]]
print(f"🔥 Total viral posts: {len(viral_posts)} / {len(posts)}")
print(f"🏆 Best platform: {all_stats[0]['platform']} (avg {all_stats[0]['avg_engagement']:,.0f} engagement)")
print("=" * 55)
````
